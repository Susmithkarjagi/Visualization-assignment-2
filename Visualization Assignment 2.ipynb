Q2. Using the tips dataset in the Plotly library, plot a box plot using Plotly express.

```python
import plotly.express as px
import seaborn as sns

# Load the tips dataset
tips = sns.load_dataset("tips")

# Create a box plot using Plotly express
fig = px.box(tips, x="day", y="total_bill", color="smoker")

# Show the plot
fig.show()
```

This code will create a box plot showing the distribution of the total_bill variable for each day of the week, separated by the smoker variable. You can modify the x and y variables to plot other variables from the dataset.

Q3. Using the tips dataset in the Plotly library, Plot a histogram for x= "sex" and y="total_bill" column in
the tips dataset. Also, use the "smoker" column with the pattern_shape parameter and the "day"
column with the color parameter.

```python
import plotly.express as px
import seaborn as sns

# Load the tips dataset
tips = sns.load_dataset("tips")

# Create a histogram using Plotly express
fig = px.histogram(tips, x="sex", y="total_bill", color="day", pattern_shape="smoker")

# Show the plot
fig.show()
```

This code will create a histogram with sex on the x-axis, total_bill on the y-axis, and bars separated by day with smoker represented by different pattern shapes. You can adjust the color_discrete_sequence parameter to modify the color scheme of the bars.

Q4. Using the iris dataset in the Plotly library, Plot a scatter matrix plot, using the "species" column for
the color parameter.

```python
import plotly.express as px
import seaborn as sns

# Load the iris dataset from seaborn
iris = sns.load_dataset("iris")

# Create a scatter matrix plot using Plotly express
fig = px.scatter_matrix(iris, dimensions=["sepal_length", "sepal_width", "petal_length", "petal_width"], color="species")

# Show the plot
fig.show()
```

This code will create a scatter matrix plot with four dimensions, sepal_length, sepal_width, petal_length, and petal_width, and points colored by the species column. The diagonal plots show the distribution of each variable, while the off-diagonal plots show the relationship between each pair of variables. You can modify the dimensions parameter to include other variables from the dataset.

Q5. What is Distplot? Using Plotly express, plot a distplot.

distplot is a function in the Seaborn library that allows you to plot a histogram along with a density plot of a single variable. It combines both the histogram and the kernel density estimation plot, providing a comprehensive view of the data distribution.

Plotly express provides the histogram function that can be used to create a similar plot. Here's an example code that shows how to create a distplot using Plotly express:

```python
import plotly.express as px
import seaborn as sns

# Load the tips dataset from Seaborn
tips = sns.load_dataset("tips")

# Create a distplot using Plotly express
fig = px.histogram(tips, x="total_bill", nbins=30, marginal="rug")

# Show the plot
fig.show()
```
